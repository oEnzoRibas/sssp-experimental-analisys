```python
import matplotlib
matplotlib.use("TkAgg")
import networkx as nx, matplotlib.pyplot as plt
%matplotlib inline
```

```python
G = nx.DiGraph()

G.add_edges_from

G.add_edge(0,1,weight=1)
G.add_edge(0,2,weight=14)
G.add_edge(0,3,weight=12)
G.add_edge(0,4,weight=11)
G.add_edge(0,5,weight=132)

pos = nx.spring_layout(G, seed=42)
nx.draw(
            G, pos,
            with_labels=True,
            node_color="#2b5c8f",  # Azul escuro elegante
            node_size=800,
        )
edge_labels = nx.get_edge_attributes(G, "weight")
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)
plt.show()
```

```python
G = nx.DiGraph()

G.add_edges_from

G.add_edge(6,1,weight=1)
G.add_edge(7,2,weight=14)
G.add_edge(8,3,weight=12)
G.add_edge(9,4,weight=11)
G.add_edge(10,5,weight=132)

G.add_edge(6,7,weight=0)
G.add_edge(7,8,weight=0)
G.add_edge(8,9,weight=0)
G.add_edge(9,10,weight=0)
G.add_edge(10,6,weight=0)

pos = nx.circular_layout(G)
nx.draw(
            G, pos,
            with_labels=True,
            node_color="#2b5c8f",  # Azul escuro elegante
            node_size=800, 
        )
edge_labels = nx.get_edge_attributes(G, "weight")
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)
plt.show()
```
